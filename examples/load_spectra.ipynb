```python
import sys, os
sys.path.insert(0, os.path.abspath("../src"))
from fulcher_analyser.spectrum import read_spectrum, plot_spectrum
%matplotlib inline
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
pth = '../echelle_spectra/out/151953_echelle_spec.txt'
pth = os.path.abspath(pth)
spec = read_spectrum(Path(pth))
```

```python
import ipywidgets as widgets
%matplotlib widget
```

```python
@widgets.interact(
    frame = list(spec.frame.data)
)

def update(frame = 1):
    """Remove old lines from plot and plot new one"""
    [artist.remove() for artist in plt.gca().lines + plt.gca().collections]
    plot_spectrum(spectrum=spec,frame=frame)
```
