# Coronal model class TEST
Check and correct Coronla modle calculations, and put them in a class for simple use.

```python
from fulcher_analyzer import coronalmodel as fcm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

# Deuterium

```python
# Load data, fit d-state with 2 Trot
arg = [150482,7,'Deuterium']
#arg = [152478,10,'Hydrogen']
isotop = arg[-1].lower()[0]

inte = fcm.read_intensities(*arg[:-1])
bp = fcm.BoltzmannPlot(inte,isotop)
bp.autofit()
```

## Figure 8, D2

```python
#plt.style.use('./ishiharastyle.mplstyle')

bp.plot_popd_paper(stylename='color',ms=4)
#ticks_ishihara()
plt.ylim([2e-2,1.2])
plt.xlim([-0.01,0.4])

#gcf().set_size_inches(figsize(8))
fcm.set_tick_size(plt.gca(),*(0.8,5,0.6,2))


plt.gcf().set_dpi(120)
plt.savefig(f"../pics/figs7-8_{bp.isotop}.svg", dpi=300,bbox_inches='tight')
```

```python
bp.print_fit_result()
bp.trotall.style.format('{:.0f}')
```

```python
como = fcm.CoronaModel(bp)
como.print_pop_shape()
como.coronal_autofit()
```

```python
#plt.style.use('./ishiharastyle.mplstyle')
como.plot_coronal_result()

axs = plt.gcf().axes
plt.sca(axs[0])

plt.savefig(f"../pics/Corona_{como.isotop}.png", dpi=300,bbox_inches='tight')
```

## Figure 10, D2

```python
#plt.style.use('./ishiharastyle.mplstyle')
como.plot_popx_paper(ms=4)

#gcf().set_size_inches(figsize(8))
fcm.set_tick_size(plt.gca(),*(0.8,5,0.6,2))
plt.gcf().set_dpi(140)

plt.savefig(f"../pics/figs9-10_{como.isotop}.svg", dpi=300,bbox_inches='tight')
```

## Figure 12, D2

```python
#plt.style.use('./ishiharastyle.mplstyle')
como.plot_paper_compare(yerr=como.yerr_flat,ms=4)

plt.gcf().set_size_inches(fcm.figsize(19,ratio=6/14))
fcm.set_tick_size(plt.gca(),*(0.8,5,0.6,2))
plt.gcf().set_dpi(140)

plt.ylim(0,0.1)

plt.savefig(f"../pics/figs11-12_{como.isotop}.svg", dpi=300,bbox_inches='tight')
```
